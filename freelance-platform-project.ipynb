# Import All Necessary Libraries

```python
import numpy as np
import pandas as pd #data cleaning 
import matplotlib.pyplot as plt # data visualization
import seaborn as sns # data visualization
import plotly.graph_objs as go
import plotly.express as px
```

```python
df=pd.read_csv('/kaggle/input/freeelance-platform-projects/Freelance Platform Projects.csv')
df
```

```python
df.info()
```

Replace null values

```python
df['Client Job Title'].replace(np.nan,'Not mentioned',inplace=True)
```

```python
df['Category Name'].unique()
```

# 1.Which are the categorys are available in freelance platform ?

```python
category_count = list(df['Category Name'].value_counts())
category_count_ls = list(df['Category Name'].value_counts().index)
fig = px.pie(values=category_count,names=category_count_ls,title='Categories in %',color_discrete_sequence=px.colors.sequential.Rainbow)
fig.update_layout(title_text='Most Category Available in freelance platform in numbers & %', title_x=0.5,height = 800)
fig.update_traces(rotation=90,textposition='inside',textinfo='label+percent+value')
```

By analyzing the distribution of projects across different categories, i could identify the most popular categories of freelance projects on the platform. 

# 2.what are the top-10  project in freelance platfrom?

```python
plt.figure(figsize = (12,8))
sns.countplot(y='Sub Category Name',order = df['Sub Category Name'].value_counts().index[0:10],data = df,palette='icefire')
plt.title('Top 10 Sub Category project as freelance platfrom')
```

**website development and Graphic design are most tranding platform to start as freelance projects **

# 3.How much budget cost for each project based on category?

```python
categories = df.groupby(['Category Name'])['Budget'].sum().reset_index()
fig = px.bar(categories,x='Category Name',y='Budget',title='Total Cost  for Each Category',
                color="Category Name",height = 700)
fig.show()
```

By analyzing the cost of different projects, identify the average cost of a freelance project on the platform. This could help clients and freelancers understand the market rate for different types of projects.

# 4. How freelancers are categorys?

```python
type_count = list(df['Experience'].value_counts())
colors = ['14213D','FCA311']
type_ls = list(df['Experience'].value_counts().index)
fig = px.pie(values=type_count,names=type_ls)
fig.update_layout(title_text='categorys based on experience  level', title_x=0.5,height = 600)
fig.update_traces(rotation=90,textposition='inside',textinfo='label+percent+value',marker=dict(colors=colors))
```

**freelancer are categories based on ther experience level **


# 5.In freelance platform which of the country are having more clients? 

```python
Client_Country = list(df['Client Country'].unique())
Client_Country_count = list(df['Client Country'].value_counts())
fig = px.pie(df,values=Client_Country_count,names=Client_Country, hole=0.5,color_discrete_sequence=px.colors.sequential.RdBu)
fig.update_layout(title_text='Num of Client from Country ', title_x=0.5)
fig.update_traces(textposition='inside',textinfo='label+percent+value')
fig.show()
```

By analyzing the location of clients and freelancers, identify the regions with the highest demand for freelance work and the regions with the most active freelancers. This could help the platform expand its reach and target new markets.

```python
df['Client Registration Date']= pd.to_datetime(df['Client Registration Date']) 
df['year_added'] = df['Client Registration Date'].dt.year
df['month_added']=df['Client Registration Date'].dt.month_name()
```

# 6.How many number of clint registered every year?

```python

df_year = df.groupby(['year_added']).agg({'Sub Category Name':'nunique'}).reset_index()
fig = px.line(df_year, x='year_added', y='Sub Category Name', markers=True,title='Overall clint registered in Each Year')
fig.show()
```

As we can see from above graph the client registration for freelance are increasesing  

```python
df_year = df.groupby(['month_added']).agg({'Sub Category Name':'nunique'}).reset_index()
fig = px.line(df_year, x='month_added', y='Sub Category Name', markers=True,title='Clint registration in Each Months w.r.t Years')
fig.show()
```

There is a peak in the month of January, which is logical since it is the month with the highest number of clints are registered.
